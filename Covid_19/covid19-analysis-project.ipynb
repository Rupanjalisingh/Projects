```python
# deleting useless columns
df = data.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
# aggregate rows by country
aggregated_data=data.groupby("Country/Region").sum()
```

```python
aggregated_data.head()
```

```python
#visualize data related to countries
aggregated_data.loc["India"].plot()
aggregated_data.loc["Germany"].plot()
aggregated_data.loc["Spain"].plot()
plt.legend()
```

```python
# Spread of corona virus in first three days
aggregated_data.loc["Spain"][:20].plot()
```

```python
# calculating the first derivative of the curve
aggregated_data.loc["India"].diff().plot()
```

```python
# Maximum infection Rate
aggregated_data.loc["India"].diff().max()
```

```python
aggregated_data = aggregated_data.apply(pd.to_numeric, errors='coerce')
```

```python
# Maximum infection Rates for all countries

countries = list(aggregated_data.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(aggregated_data.loc[c].diff().max())

aggregated_data["Max_infection_rates"] = max_infection_rates
```

```python
aggregated_data
```

```python
#create new dataframe
covid_data = pd.DataFrame(aggregated_data["Max_infection_rates"])
```
